```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

# 1) Interpolation

```python
x = np.linspace(0, 10, 10)
y = x**2
fig , ax = plt.subplots()
ax.scatter(x, y)
```

```python
f = sp.interpolate.interp1d(x, y)

new_x = np.linspace(0, 10, 30)
result = f(new_x)
fig , ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(new_x, result, color='red')
```

```python
x= np.linspace(0, 2, 100)
y = 1/3 * x ** 3 - 3/5 * x ** 2 + np.random.randn(x.shape[0])/20
fig , ax = plt.subplots()
ax.scatter(x, y)
```

```python
def f(x, a, b, c, d):
    return a * x ** 3 + b * x ** 2 + c * x + d
```

```python
params , params_cov = sp.optimize.curve_fit(f , x , y)
```

```python
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), color='red')
```

# Minimisation

```python
def f(x):
    return x**2 + 15 * np.sin(x)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x))
```

```python
x0 = -5
result = sp.optimize.minimize(f, x0=x0).x
```

```python
fig , ax = plt.subplots()
ax.plot(x, f(x))
ax.scatter(result, f(result), color='red')
ax.scatter(x0, f(x0), color='green')
plt.show()
```

```python
def f (x):
    return np.sin(x[0]) + np.cos(x[0]+x[1])*np.cos(x[0])
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

x, y = np.meshgrid(x, y)
plt.contour(x, y, f([x, y]), 20)

x0 = np.zeros((2, 1))
plt.scatter(x0[0], x0[1], color='red')
sp.optimize.minimize(f, x0=x0)
```
