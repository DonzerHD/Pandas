```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

# 1) Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(x, y , label='quadratique')
plt.plot(x, x**3, label = 'cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r--')
plt.subplot(2,1,2)
plt.plot(x, x**3, 'g*-')
```

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('figure 1')
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.legend()
```

POO matplotlib

```python
fig , ax = plt.subplots(2,2, sharex=True)
ax[0, 1].plot(x, y)
ax[1, 1].plot(x, np.sin(x), label='sinus')
plt.show()
```

Exo

```python
dataset = {f"experience  {i}": np.random.randn(100) for i in range(4)}
dataset
```

```python
def graphique(dataset):
    fig, ax = plt.subplots(4,1 , sharex=True)
    i = 0
    for k, v in dataset.items():
        ax[i].plot(v, label=k)
        ax[i].legend()
        i += 1
    plt.show()
graphique(dataset)
```
