# Exercice 3
Sur l'image ci dessous, effectuer un slicing pour ne garder que la moitiÃ© de l'image (en son centre) et remplacer tous les pixels > 150 par des pixels = 255

```python
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
face.shape
```

```python
face_half = face[2:500, 400:900]

# Remplacer les pixels > 150 par des pixels = 255
face_half[face_half > 150] = 255

# Afficher l'image
plt.imshow(face_half, cmap=plt.cm.gray)
plt.show()
```
