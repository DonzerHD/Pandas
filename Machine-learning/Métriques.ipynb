```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

```python
y = np.array([1, 2 , 2 , 3 , 5 , 2])
y_pred = np.array([5 , 2 , 2 , 5 , 7 , 1000])

print('mean_absolute_error : ', mean_absolute_error(y, y_pred))
print('RMSE : ', np.sqrt(mean_squared_error(y, y_pred)))
print('median_absolute_error : ', median_absolute_error(y, y_pred))
```

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('housing.csv')
df.corr()
# seaborn corr
sns.heatmap(df.corr(), annot=True)
```

# Test Californie maison

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LinearRegression

X = df[['median_income' , 'total_rooms' , 'housing_median_age' , 'population']]
y = df['median_house_value']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print('R2 : ', r2_score(y, y_pred))

plt.scatter(X.iloc[:, 0], y,  alpha=0.8, label='y')
plt.scatter(X.iloc[:, 0], y_pred ,  alpha=0.8, label='y_pred')
plt.legend()
```
